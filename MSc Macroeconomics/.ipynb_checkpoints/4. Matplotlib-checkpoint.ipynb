## Import

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Specification

```python
%matplotlib inline
```

```python
x = np.linspace(-10, 10, 50)
y = x ** 2

plt.figure(figsize=(15,8))
plt.plot(x, y, c = 'r', ls = '-', lw = 3, marker = 'o', markersize = 15,
         markeredgecolor = 'b', markerfacecolor = 'm', label = '$y = x^2$')

plt.legend(loc = 'upper center', fontsize = 15)
plt.grid(c = 'k', ls = '--', alpha = 0.8)

plt.xlabel('X range', fontsize = 15, fontweight = 2, fontname = 'Times New Roman')
plt.ylabel('Y range', fontsize = 15, fontweight = 2, fontname = 'Times New Roman')
plt.title('Function $Y=X^2$ in range (-10,10)', fontsize = 20, fontweight = 2, fontname = 'Times New Roman', 
         c = 'darkblue')

#plt.text(-5, 60, 'first text\nin figure', fontsize = 15, fontweight = 2, fontname = 'Times New Roman')

#plt.xlim(left = -4, right = 8)
#plt.ylim(top = 55, bottom = 10)

plt.axhline(40, c = 'k', lw = 2, ls = '--' )
plt.axvline(0, c = 'g', lw = 2, ls = ':' )

#plt.savefig('text_1.jpg', dpi = 700)

plt.show()
```

```python
x = np.linspace(-10,10, 50)
y = x ** 2

plt.figure(figsize=(15,8))
plt.plot(x, 10*y,c = 'r', ls = '-', lw = 3, label='$y=x^2$')
plt.plot(x, x**3, c = 'b', ls = '--', lw = 3, label='$y=x^3$')
plt.plot(x, 250 * np.sin(x), c = 'g', ls = '--', lw = 3, label='$y=sin(x)$')

plt.legend(fontsize = 15)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.subplot(2, 3, 1)
plt.plot(x, x**2)

plt.subplot(2,3,2)
plt.plot(x, x**3)
plt.grid()

plt.subplot(2,3,3)
plt.plot(x, np.sin(x))

plt.subplot(2,3,5)
plt.plot(x, np.tan(x))

plt.show()
```

## Bar

```python
plt.figure(figsize=(15,8))

plt.bar([1,3,5,7], [5,8,2,6], label = 'one')
plt.bar([2,4,6,8], [1,4,8,3], label = 'two')

plt.grid(axis = 'y', ls = '--')

plt.legend(fontsize = 15)
plt.show()
```

## Scatter

```python
x = np.linspace(0, 10, 100)
e = np.random.rand(100) * 100
y = x + e

scale = np.random.rand(100) * 1000
color = np.random.rand(100) * 10

plt.figure(figsize=(15,8))
plt.scatter(x, y, s = scale, c = color)

plt.grid()

plt.colorbar().set_label('color bar', fontsize =15)

plt.show()
```

## Histogram

<img stc = https://wikimedia.org/api/rest_v1/media/math/render/svg/00cb9b2c9b866378626bcfa45c86a6de2f2b2e40 >

```python
data = np.random.normal(0,1, 100000)

x = np.linspace(-4,4, 100)
y = 1 / np.sqrt(2 * np.pi) * np.exp(-x**2/2)

plt.figure(figsize=(15,8))
plt.plot(x, y, lw = 4, c = 'k')
plt.hist(data, bins=70, density=1)

plt.show()
```
